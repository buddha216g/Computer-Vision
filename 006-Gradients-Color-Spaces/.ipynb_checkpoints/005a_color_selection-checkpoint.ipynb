### Import Packages

```python
#importing some useful packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline
```

## Read in an Image

```python
import os
print(os.listdir("../006-Gradients-Color-Spaces"))
```

```python
#reading in an image
image = mpimg.imread('../006-Gradients-Color-Spaces/Yellow_Lane_Straight.jpg')

#printing out some stats and plotting
print('This image is:', type(image), 'with dimensions:', image.shape)
plt.imshow(image)  # if you wanted to show a single color channel image called 'gray', for example, call as plt.imshow(gray, cmap='gray')
```

```python
# extract R channel
R_channel = image[:,:,0]

# create empty image with same shape as that of src image
R_img = np.zeros(image.shape)

#assign the red channel of src to empty image
R_img[:,:,0] = R_channel

plt.imshow(R_channel)
```

```python
# extract G channel
G_channel = image[:,:,1]

# create empty image with same shape as that of src image
G_img = np.zeros(image.shape)

#assign the red channel of src to empty image
G_img[:,:,1] = G_channel

plt.imshow(G_channel)
```

```python
# extract B channel
B_channel = image[:,:,2]

# create empty image with same shape as that of src image
B_img = np.zeros(image.shape)

#assign the red channel of src to empty image
B_img[:,:,2] = B_channel

plt.imshow(B_channel)
```

```python
# Other color channels
hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
H = hls[:,:,0]
S = hls[:,:,2]
L = hls[:,:,1]
```

```python
# extract h channel
H_channel = hls[:,:,0]

# create empty image with same shape as that of src image
H_img = np.zeros(image.shape)

#assign the H channel of src to empty image
H_img[:,:,0] = H_channel

plt.imshow(H_channel)

#cv2.imwrite('../006-Gradients-Color-Spaces/H_channel.jpg',H_img) 
```

```python
# extract L channel
L_channel = hls[:,:,1]

# create empty image with same shape as that of src image
L_img = np.zeros(image.shape)

#assign the red channel of src to empty image
L_img[:,:,1] = L_channel

plt.imshow(L_channel)

#cv2.imwrite('../006-Gradients-Color-Spaces/L_channel.jpg',L_img)
```

```python
# extract S channel
S_channel = hls[:,:,2]

# create empty image with same shape as that of src image
S_img = np.zeros(image.shape)

#assign the red channel of src to empty image
S_img[:,:,2] = S_channel

plt.imshow(S_channel)

#cv2.imwrite('../006-Gradients-Color-Spaces/S_channel.jpg',S_img)
```

```python
f, ((ax1, ax2,ax3), (ax4,ax5,ax6)) = plt.subplots(2,3,figsize=(22,8),sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0.2, 'wspace': 0.2})
ax1.imshow(R_channel)
ax1.set_title('R Channel', fontsize=20)
ax2.imshow(G_channel)
ax2.set_title('G Channel', fontsize=20)
ax3.imshow(B_channel)
ax3.set_title('B Channel', fontsize=20)
ax4.imshow(H_channel)
ax4.set_title('H Channel', fontsize=20)
ax5.imshow(L_channel)
ax5.set_title('L Channel', fontsize=20)
ax6.imshow(S_channel)
ax6.set_title('S Channel', fontsize=20)
```

```python
f.savefig('../006-Gradients-Color-Spaces/Colour_channels.jpg')
plt.close(f)
```
