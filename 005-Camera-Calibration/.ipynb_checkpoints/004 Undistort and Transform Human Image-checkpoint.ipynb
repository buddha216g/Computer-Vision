```python
import pickle
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline


# Read in an image
orig_image = cv2.imread('test1.jpg')
img = np.copy(orig_image)
print(img.shape)
plt.imshow(img)
```

```python
src = np.float32([[800, 4100],[2650, 4050], [2580, 2220],[750, 2220]])

plt.imshow(img)
plt.plot( 750, 2220,'.')
plt.plot( 2580, 2220,'.')
plt.plot( 800, 4100,'.')
plt.plot( 2650, 4050,'.')
```

```python
dst = np.float32([[750, 4100],[2580, 4100], [2580, 2220],[750, 2220]])
plt.imshow(img)
plt.plot( 750, 2220,'*')
plt.plot( 2580, 2220,'*')
plt.plot( 750, 4100,'*')
plt.plot( 2580, 4100,'*')
```

```python
img_size = (img.shape[1],img.shape[0])
print(img_size)
M = cv2.getPerspectiveTransform(src, dst)
warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(img)
ax1.set_title('Original Image', fontsize=50)
ax2.imshow(warped)
#ax2.imshow(warped)
ax2.set_title('Warped Image', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```
